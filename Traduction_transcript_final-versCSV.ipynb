```python
from youtube_transcript_api import YouTubeTranscriptApi
from deep_translator import GoogleTranslator
import unicodedata
import csv
import chardet
from datetime import datetime
import time
import glob
import string
import pandas as pd
from pytube import YouTube
import nltk
from detect_delimiter import detect
from mtranslate import translate
import re
from langdetect import detect
from itertools import zip_longest
```

```python
characters = "/_\[]@|~{}@!?"
def supp_special_carac (srt):
  all_trans=""
  for i in srt:
    
    if ((i['text'].lower() == "[musique]")or (i['text'].lower() == "[music]")) :
      i['text'] = " "
                
    if ((i['text'].lower() == "[applause]") or (i['text'].lower() == "[applaudissements]")):
      i['text'] = ""
    for x in range(len(characters)):
      i["text"] = i["text"] .replace(characters[x],"")
    a=re.sub("\d\d\:\d\d|\d\d\\s:\s\d\d|\d\d\\s:\d\d|\d\d\:\s\d\d","",i["text"])
    
    all_trans= all_trans+" "+ a
    
  
  return(all_trans)
```
